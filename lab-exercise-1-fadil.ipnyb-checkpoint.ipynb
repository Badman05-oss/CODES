# Excercise 2
#### Python program to calculate the sum of three given numbers , If the values are equal then return thrice of their sum

 {pictural representation}


```python
def sum_thrice(x,y,z):
    sum = x + y + z
    if x == y == z:
        sum = sum * 3
        return sum
    print(sum)
    
print(sum_thrice(1,2,3))

print(sum_thrice(3,3,3)) 
```

# Excercise 3
### Python program which returns true if the two given integer values are equal or their sum or difference is 5

 {pictorial representation}

```python
def number(x,y):
    if x==y :
        return True
    elif (x + y) == 5:
        return True
    elif (x - y) == 5:
        return True
    else:
        return False
print(number(7,2))
print(number(3,2))
print(number(2,2))
```

# Excercise 5
### python function that takes a positive integer and returns the sum of the cube of all the positive integer smaller than the specified number.

 {Pictorial representation}

```python
def value(n):
    n = n-1
    total = 0
    while n > 0:
        total += n * n * n
        n= n-1
        
    return total
print("Sum of cubes: ", value(6))
```
