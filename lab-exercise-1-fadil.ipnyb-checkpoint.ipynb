#### python program to get the differemce between a given number and 17 , if the number is greater than 17 return the double the absolute difference 

 {The pictorial representation}

```python
number = int(input("Pick a random number"))
new_number = number - 17
if number>17:
    print (2*abs(new_number))
else:
    print(abs(new_number))
```
